# Ejercicios con Tkinter

A continuación se presentan tres ejemplos prácticos para ejercitar el uso de Tkinter en Python, incluyendo ventanas, botones y listas desplegables.

## Ejemplo 1: Ventana con botón que cambia el texto de una etiqueta

```python
import tkinter as tk

def cambiar_texto():
    etiqueta.config(text="Texto cambiado")

ventana = tk.Tk()
ventana.title("Ejemplo 1")

etiqueta = tk.Label(ventana, text="Texto original")
etiqueta.pack()

boton = tk.Button(ventana, text="Cambiar texto", command=cambiar_texto)
boton.pack()

ventana.mainloop()
```

## Ejemplo 2: Entrada de texto y botón para mostrar mensaje

```python
import tkinter as tk

def mostrar():
    etiqueta.config(text=f"Ingresaste: {entrada.get()}")

ventana = tk.Tk()
ventana.title("Ejemplo 2")

entrada = tk.Entry(ventana)
entrada.pack()

boton = tk.Button(ventana, text="Mostrar", command=mostrar)
boton.pack()

etiqueta = tk.Label(ventana, text="Esperando entrada")
etiqueta.pack()

ventana.mainloop()
```

## Ejemplo 3: Lista desplegable para seleccionar una opción

```python
import tkinter as tk

def seleccionar(opcion):
    etiqueta.config(text=f"Seleccionaste: {opcion}")

ventana = tk.Tk()
ventana.title("Ejemplo 3")

opciones = ["Rojo", "Verde", "Azul"]
variable = tk.StringVar(ventana)
variable.set(opciones[0])

menu = tk.OptionMenu(ventana, variable, *opciones, command=seleccionar)
menu.pack()

etiqueta = tk.Label(ventana, text="Elige una opción")
etiqueta.pack()

ventana.mainloop()
```
