# 📌 Funciones Más Usadas en NumPy y Pandas

Este notebook muestra ejemplos de las funciones más utilizadas en las bibliotecas NumPy y Pandas.


## 🔹 1. Funciones Más Usadas de NumPy
NumPy es una biblioteca para cálculos numéricos en Python. A continuación, se muestran 10 de sus funciones más utilizadas con ejemplos.


```python
import numpy as np

# 1. Crear un arreglo
arr = np.array([1, 2, 3, 4, 5])
print(arr)

# 2. Generar un arreglo de ceros
print(np.zeros(5))

# 3. Generar un arreglo de unos
print(np.ones(5))

# 4. Crear un rango de números
print(np.arange(1, 10, 2))

# 5. Generar números aleatorios
print(np.random.rand(3))

# 6. Calcular la media
print(np.mean(arr))

# 7. Calcular la mediana
print(np.median(arr))

# 8. Calcular la desviación estándar
print(np.std(arr))

# 9. Encontrar el valor máximo
print(np.max(arr))

# 10. Encontrar el valor mínimo
print(np.min(arr))
```

## 🔹 2. Funciones Más Usadas de Pandas
Pandas es una biblioteca para el análisis de datos en Python. A continuación, se muestran 10 de sus funciones más utilizadas con ejemplos.


```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
print(df)

# 1. Ver las primeras filas
print(df.head())

# 2. Ver las últimas filas
print(df.tail())

# 3. Obtener información del DataFrame
print(df.info())

# 4. Obtener estadísticas descriptivas
print(df.describe())

# 5. Seleccionar una columna
print(df['A'])

# 6. Filtrar datos
print(df[df['A'] > 1])

# 7. Agregar una nueva columna
df['C'] = df['A'] + df['B']
print(df)

# 8. Eliminar una columna
df = df.drop(columns=['C'])
print(df)

# 9. Reemplazar valores en una columna
df['A'] = df['A'].replace(1, 10)
print(df)

# 10. Ordenar por una columna
print(df.sort_values(by='B', ascending=False))
```

## 🔥 Conclusión
- NumPy es ideal para cálculos numéricos rápidos y eficientes.
- Pandas facilita la manipulación y análisis de datos estructurados.
- Ambas bibliotecas son esenciales para el análisis de datos y aprendizaje automático en Python.
